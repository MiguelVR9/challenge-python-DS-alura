### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
```

#1. Análisis de facturación



## *Ingresos totales de las cuatro tiendas*

```python
# Ingreso total de la tienda 1
ingreso_t1 = tienda['Precio'].sum()
print(f"Ingreso total de la tienda 1: ${ingreso_t1:,.2f}")
```

```python
# Ingreso total de la tienda 2
ingreso_t2 = tienda2['Precio'].sum()
print(f"Ingreso total de la tienda 2: ${ingreso_t2:,.2f}")
```

```python
# Ingreso total de la tienda 3
ingreso_t3 = tienda3['Precio'].sum()
print(f"Ingreso total de la tienda 3: ${ingreso_t3:,.2f}")
```

```python
# Ingreso total de la tienda 4
ingreso_t4 = tienda4['Precio'].sum()
print(f"Ingreso total de la tienda 4: ${ingreso_t4:,.2f}")
```

```python
ingresos = {
    "Tienda 1": ingreso_t1,
    "Tienda 2": ingreso_t2,
    "Tienda 3": ingreso_t3,
    "Tienda 4": ingreso_t4
}

#Creación de la figura y ejes
fig, axs = plt.subplots()
g_barras1= axs.bar(ingresos.keys(), ingresos.values(), color = 'lightgreen')
axs.bar_label(g_barras1, labels=[f'{v:,.2f}' for v in ingresos.values()])
axs.set_ylim(1e9, 1.2e9)
axs.set_title("Ingresos totales de las cuatro tiendas")
axs.set_xlabel("Tiendas")
axs.set_ylabel("Ingresos")

plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Valoración media de la tienda 1
valoracion_t1 = tienda['Calificación'].mean()
print(f"Valoración media de la tienda 1: {valoracion_t1:.2f}")
```

```python
# Valoración media de la tienda 2
valoracion_t2 = tienda2['Calificación'].mean()
print(f"Valoración media de la tienda 2: {valoracion_t2:.2f}")
```

```python
# Valoración media de la tienda 3
valoracion_t3 = tienda3['Calificación'].mean()
print(f"Valoración media de la tienda 3: {valoracion_t3:.2f}")
```

```python
# Valoración media de la tienda 4
valoracion_t4 = tienda4['Calificación'].mean()
print(f"Valoración media de la tienda 4: {valoracion_t4:.2f}")
```

```python
#Gráfico de la valoración media de las cuatro tiendas
valoracion_media = {
    "Tienda 1": valoracion_t1,
    "Tienda 2": valoracion_t2,
    "Tienda 3": valoracion_t3,
    "Tienda 4": valoracion_t4
}

# Colores distintos para cada punto
colores = ['blue', 'green', 'yellow', 'red']
# Crear figura
plt.figure(figsize=(8,5))
plt.scatter(valoracion_media.keys(), valoracion_media.values(), color=colores, marker='*', s=200)
# Mostrar el valor de cada punto al lado
for x, y in zip(valoracion_media.values(), valoracion_media.keys()):
    plt.text(y, x+0.1, f'{x:,.2f}', va='bottom', ha='center',fontsize=10)

plt.ylim(1, 5)
plt.title("Valoración media de las cuatro tiendas")
plt.xlabel("Tiendas")
plt.ylabel("Valoración media")
plt.title("Valoración media de las cuatro tiendas")
plt.show()

```

# 5. Envío promedio por tienda

```python
#Costo de envío promedio de la tienda 1
costo_envio_prom_t1 = tienda['Costo de envío'].mean()
print(f"El costo de envío promedio de la tienda 1 es: {float(costo_envio_prom_t1):,.2f}")
```

```python
#Costo de envío promedio de la tienda 2
costo_envio_prom_t2 = tienda2['Costo de envío'].mean()
print(f"El costo de envío promedio de la tienda 2 es: {float(costo_envio_prom_t2):,.2f}")
```

```python
#Costo de envío promedio de la tienda 3
costo_envio_prom_t3 = tienda3['Costo de envío'].mean()
print(f"El costo de envío promedio de la tienda 3 es: {float(costo_envio_prom_t3):,.2f}")
```

```python
#Costo de envío promedio de la tienda 4
costo_envio_prom_t4 = tienda4['Costo de envío'].mean()
print(f"El costo de envío promedio de la tienda 4 es: {float(costo_envio_prom_t4):,.2f}")
```

```python
#Gráfica del costo de envío promedio de las 4 tiendas
costo_envio_prom = {
    "Tienda 1": costo_envio_prom_t1,
    "Tienda 2": costo_envio_prom_t2,
    "Tienda 3": costo_envio_prom_t3,
    "Tienda 4": costo_envio_prom_t4
}
#Creación de la figura y ejes
fig, axs = plt.subplots()
g_barras2= axs.bar(costo_envio_prom.keys(), costo_envio_prom.values(), color = 'orange')
axs.bar_label(g_barras2, fmt='%.2f')
axs.set_title("Costo de envío promedio de las 4 tiendas")
axs.set_xlabel("Tiendas")
axs.set_ylabel("Costo de envío promedio")
```
