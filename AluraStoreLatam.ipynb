#1. Análisis de facturación



## *Ingresos totales de las cuatro tiendas*

```python
# Ingreso total de la tienda 1
ingreso_t1 = tienda['Precio'].sum()
print(f"Ingreso total de la tienda 1: ${ingreso_t1:,.2f}")
```

```python
# Ingreso total de la tienda 2
ingreso_t2 = tienda2['Precio'].sum()
print(f"Ingreso total de la tienda 2: ${ingreso_t2:,.2f}")
```

```python
# Ingreso total de la tienda 3
ingreso_t3 = tienda3['Precio'].sum()
print(f"Ingreso total de la tienda 3: ${ingreso_t3:,.2f}")
```

```python
# Ingreso total de la tienda 4
ingreso_t4 = tienda4['Precio'].sum()
print(f"Ingreso total de la tienda 4: ${ingreso_t4:,.2f}")
```

```python
ingresos = {
    "Tienda 1": ingreso_t1,
    "Tienda 2": ingreso_t2,
    "Tienda 3": ingreso_t3,
    "Tienda 4": ingreso_t4
}

#Creación de la figura y ejes
fig, axs = plt.subplots()
g_barras1= axs.bar(ingresos.keys(), ingresos.values(), color = 'lightgreen')
axs.bar_label(g_barras1, labels=[f'{v:,.2f}' for v in ingresos.values()])
axs.set_ylim(1e9, 1.2e9)
axs.set_title("Ingresos totales de las cuatro tiendas")
axs.set_xlabel("Tiendas")
axs.set_ylabel("Ingresos")
plt.savefig("ingresos.png")
plt.show()
```

# 3. Calificación promedio de la tienda

```python
#Gráfico de la valoración media de las cuatro tiendas
valoracion_media = {
    "Tienda 1": valoracion_t1,
    "Tienda 2": valoracion_t2,
    "Tienda 3": valoracion_t3,
    "Tienda 4": valoracion_t4
}

# Colores distintos para cada punto
colores = ['blue', 'green', 'yellow', 'red']
# Crear figura
plt.figure(figsize=(8,5))
plt.scatter(valoracion_media.keys(), valoracion_media.values(), color=colores, marker='*', s=200)
# Mostrar el valor de cada punto al lado
for x, y in zip(valoracion_media.values(), valoracion_media.keys()):
    plt.text(y, x+0.1, f'{x:,.2f}', va='bottom', ha='center',fontsize=10)

plt.ylim(1, 5)
plt.title("Valoración media de las cuatro tiendas")
plt.xlabel("Tiendas")
plt.ylabel("Valoración media")
plt.title("Valoración media de las cuatro tiendas")
plt.show()

```

## Conclusión

### Recomendación de la tiende que se debe vender

Los cinco puntos analizados en este proyecto muestran que algunas tiendas destacan en ciertas métricas mientras que otras presentan debilidades. Por ello, en la toma de decisión se dio prioridad a las métricas con mayor impacto financiero.

Con base en esto, se recomienda que el Sr. Juan venda la **Tienda 4**, ya que presenta el desempeño global más bajo. Aunque tiene buenas ventas en algunas categorías específicas (Juguetes y Artículos para el hogar) y un costo de envío bajo, sus ingresos totales son los menores, sus ventas en las categorías más demandadas y estratégicas (Electrónicos y Muebles) son insuficientes, y los productos más vendidos no coinciden con estas categorías clave, lo que limita su eficiencia y potencial de crecimiento. Su calificación promedio es aceptable, pero no compensa las deficiencias en ingresos y cobertura de categorías. Por estas razones, se recomienda al Sr. Juan vender la **Tienda 4**, manteniendo las demás tiendas con mejor rendimiento y mayor alineación con la demanda de los clientes.
