# 2. Ventas por categoría

```python
# Categorías de la tienda 1 ordenada de la más pupular a la menos popular por cantidad de productos vendidos
ventas_categorias_t1 = tienda.groupby('Categoría del Producto').size().sort_values(ascending=False)
ventas_categorias_t1
```
